```python
#딥러닝 과제
#다음주 월요일까지
#데이터 수집해오기.?
#행50이상 열4이상
#예측 혹은 분류하고 싶은 대상의 데이터(수치 기반)를 수집 혹은 모아오기
#독립변수와 종속변수가 존재해야함
#데이터 수집 방법은 자율적
```

```python
class Y_pred():
  def __init__(self,theta_0,theta_1):
    self.theta_0=theta_0
    self.theta_1=theta_1
```

```python
def out_y_hat(self,input_x):
  y_hat_row=[]
  for i in range(len(input_x)):
    y_hat=self.theta_0+self.theta_1*input_x[i]
    y_hat_row.append(y_hat)
  return y_hat_row

Y_pred.y_hat=out_y_hat
```

```python
Y_pred_C = Y_pred(theta_0=1,theta_1=2)
x=[1000]
Y_pred_C.y_hat(x)
```

```python
#목적함수 ///// mse는 1/2이 없다. 1/2는 미분할때 식이 더 깔끔하게 하는 역할
#목적함수 : 정량적 지표를 높일 것인지, 낮출것인지에 대하여 목적을 가지고 있는 함수
#손실함수 : 신경망의 성능을 측정하고자 출력한 정략적 지표
#목적함수가 좀 더 큰 바운더리, 그 안에 손실함수, 안에 mse,,sse,,등등
#목적함수와 손실함수는 거의 같은 개념을 의미한다..
```

```python
import numpy as np
y_label_total=[374,385,375,401]
y_hat_total=[1161, 1401, 1621, 1681]

diff_row=[]
for i in range(len(y_label_total)):
  diff = y_label_total[i] - y_hat_total[i]
  diff_row.append(diff)
  square=np.square(diff_row)
  sse = 1/2*(np.sum(square))

print('Diff {} \nSquare : {}\nSSE : {}'.format(diff_row,square,sse))
```

* eta / learning_rate(lr) / 값의 갱신되는 보폭(폭)을 제어하는 역할 / 기본값으로 0.001 / 양의값만을 가짐, 음수는 위험 /  

```python
# 학습률 : 1
# x값 : 3
# 갱신횟수 : 4
def g(x,learning_rate):
  print("Now x value : {}".format(x))
  epoch_1_x = x - learning_rate*((2*x)-2)
  print("Epoch 1 - x : {}".format(epoch_1_x))

  epoch_2_x = epoch_1_x - learning_rate*((2*epoch_1_x)-2)
  print("Epoch 1 - x : {}".format(epoch_2_x))

  epoch_3_x = epoch_2_x - learning_rate*((2*epoch_2_x)-2)
  print("Epoch 1 - x : {}".format(epoch_3_x)) 

  epoch_4_x = epoch_3_x - learning_rate*((2*epoch_3_x)-2)
  print("Epoch 1 - x : {}".format(epoch_4_x)) 
```

```python
g(x=3, learning_rate=1)
```

```python
g(x=3, learning_rate=0.001)
```

```python
g(x=3, learning_rate=0.01)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def g(learning_rate,epoch_count,random_value_bool=False,x=3):
  if random_value_bool:
    x=np.random.normal(0,1,size=1)
  print("Now x value: {}".format(x))
  print("Learning Rate : {}".format(learning_rate))
  x_row=[]
  for i in range(epoch_count):
    x = x - learning_rate*((2*x)-2)
    print('Epoch : {} / x : {}'.format(i+1, np.round(x,3)))
    x_row.append(x)
  
  return x_row
```

```python
result = g(learning_rate=0.1,epoch_count=20,random_value_bool=True)
```

```python
plt.plot(result,'--o',color='green')
plt.xlabel('Epoch')
plt.ylabel('X value')
plt.grid()
plt.show()
```

```python
data=pd.read_csv(drive_path + 'bmi_500.csv')

X=data.loc[:,'Height':'Weight']  #숫자로 가져오고 싶으면 iloc, 문자면 loc
y=data.loc[:,'Label']
print(X.shape)
print(y.shape)

X_train = X.iloc[:350, :]
X_test = X.iloc[350:, :]
y_train = y.iloc[:350]
y_test = y.iloc[350:]

bmi_model = KNeighborsClassifier(n_neighbors=10)
bmi_model.fit(X_train, y_train)
pre=bmi_model.predict(X_test)
metrics.accuracy_score(y_test,pre)
```
