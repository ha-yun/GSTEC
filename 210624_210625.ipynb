```python
#add_subplot()

fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax1.plot([1,2,3],[4,10,6])
ax2=fig.add_subplot(2,1,2)
ax2.plot([1,10,100],[1,100,10])
plt.show()
```

```python
fig,ax=plt.subplots(2,1)
plt.show()
```

```python
fig,ax=plt.subplots(2,1)
ax[0].plot([1,2,3],[4,10,6])
ax[1].plot([1,10,100],[1,100,10])
plt.show()
```

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot([1,2,3],[4,10,6])
ax[0,1].plot([1,10,100],[1,100,10])
ax[1,0].plot([4,10,6],[1,2,3])
ax[1,1].plot([1,100,10],[1,10,100])
plt.show()
```

<p>matplotlib API</p>
API= application programming interface
응용 프로그램에서 사용할 수 있도록 운영체제나 프로그래밍 언어가 제공하는 기능을 제어할 수 있게 만든 인터페이스
<p>Matplotlib에는 주로 Pyplot API와 Object-Oriented API로 작동</p>

```python
plt.plot([0,1,2],[4,10,6])
plt.show()
```

```python
plt.plot([1,2,3],[1,2,3])
plt.plot([1,2,3],[1,4,9])
plt.plot([1,2,3],[1,5,10])
plt.show()
```

```python
#Object-Oriented API
fig,ax=plt.subplots(2,2)
ax[0,0].plot([1,2,3],[4,10,6])
ax[0,1].plot([1,10,100],[1,100,10])
ax[1,0].plot([4,10,6],[1,2,3])
ax[1,1].plot([1,100,10],[1,10,100])
plt.show()
```

27.그래프 그리기

```python
import numpy as np
```

```python
#배열만들기 linspace()
x=np.linspace(0,15,1000)
fig,ax=plt.subplots(2,1)
ax[0].plot(x,np.cos(x))
ax[1].plot(x,np.sin(x))
plt.show()
```

```python
x=np.linspace(0,15,1000)
plt.plot(x,np.cos(x),color='red',label='cos')
plt.plot(x,np.sin(x),color='blue',label='sin')
plt.title('Cosine & Sine')   #title() : 그래프 위에 제목 추가
plt.xlabel('x label')     #xlabel : x축아래에 축 이름 추가
plt.ylabel('y label')     #ylabel : y축왼쪽에 축 이름 추가
plt.legend()             #legend() : 그래프에 범례 추가
plt.show()
```

```python
x=np.linspace(0,15,1000)
plt.plot(x,np.cos(x),color='red',label='cos')
plt.plot(x,np.sin(x),color='blue',label='sin')
plt.title('Cosine & Sine')
plt.xlabel('x label')
plt.ylabel('y label')  
plt.legend(loc='upper right')    #범례 위치 변경 ==plt.legend(loc=1) 등등등
```

```python
#그래프 색 변경
x=np.linspace(0,15,1000)
plt.plot(x,np.cos(x),color='c',label='cyan cos')    #color=cyan
plt.plot(x,np.sin(x),color='m',label='magenta sin') #color=magenta
plt.title('Color')
plt.xlabel('x label')
plt.ylabel('y label')  
plt.legend(loc=4)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font',family='NanumBarunGothic')
```

```python
plt.plot([0,1,2],[4,10,6])
plt.show()
```

```python
x=np.arange(0,10,1)
plt.scatter(x,x+1)
plt.title('Scatter Plot')
plt.show()
```

```python
data=[10,13,20,16]
x=range(len(data))
plt.bar(x,data)
plt.title('Bar Chart')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
yerr=[(1,3,5,5,10),(3,3,7,5,10)]
plt.errorbar(x,y,yerr)
plt.title("Error Bar Chart")
plt.show()
```

```python
x=np.random.normal(0,1,size=5000)
plt.hist(x)
plt.title("Histogram")
plt.xlabel("x가 속하는 구간")
plt.ylabel("구간에 속하는 x의 개수")
plt.show()
```

```python
x=np.random.normal(0,1,size=5000)
plt.hist(x,bins=100,density=True)
plt.title('Histogram')
plt.xlabel('x가 속하는 구간')
plt.ylabel('각 bin에 속하는 개수/전체 개수(정규화)')
plt.show()
```

```python
#Pie Chart
data=[60,30,10,30,70]
categories=['Hip Hop','Jazz','Blues','Rock','Pop']
plt.pie(data,labels=categories,autopct='%0.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Box Plot
x=np.random.normal(100,20,300)
plt.boxplot(x)
plt.title('Box Plot')
plt.show()
```
