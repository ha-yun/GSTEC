<h1>HTML Tag </h1>
1. HTML tag  : html문서라면 꼭 써줘야 한다.<br>
**  !DocTYPE html <br>
**  html , /html </p>
2. Head tag : 내용물에 대한 추가 정보 <br>
3. body tag : 내용물의 모든 내용 <br>
4. title tag : 페이지 이름(주소창) <br>
5. p tag : paragraph <br>
6. link tag : index.html과 codelion.css파일을 연결하는데 필요한 코드로 head tag안에 쓴다. <br>
7. div tag : division 여러 태그들을 묶어서 css를 함께 꾸며주기 위해 쓰이는 태그
(같은 태그지만 따로 꾸미고 싶을 땐 class 사용)<br>
8. section tag : division tag와 완전히 똑같음, html요소들을 꾸미기 위해 묶어주는 역할<br>
9. article tag ; division, section과 같음<br>
(div,section,article tag를 적절하게 분배하여 난잡하지 않은 코드를 만든다.)<br>
10. img src="" : 이미지 삽입<br>
11. h1,h2,h3.. tag<br>
+ 로렘 입숨 lorem ipsum

```python
'''
##index.html 파일

<!DOCTYPE html>
<html>
  <head>
    <meta charset="UTF-8">
    <title>현이력서</title>
    <link rel="stylesheet" href="codelion.css">
  </head>

  <body>
    <div class="mainbox">
        <div class="title-box">
            <p class="big-font">what</p>
            <p>About</p>
        </div>
        <footer>다섯번째 계절</footer>

        <section>
            <h1>min9</h1>
            <p class="small-font">she want pizza</p>
            <article>
                <h2>ABOUT HER</h2>
                <p class="about-me-text"> she is my friend </p>
                <p class="small-font">HTML/CSS programmar</p>
            </article>
        </section>

        <section>
        <div class="float-wrap">
            <h2>Experience</h2>
            <p class="title-text">Awesome day</p>
            <p class="title-date">2020-06-25</p>
        </div>
          <p class="desc-text">Front Web Developer</p>
          <p class="desc-subtext">HTML/CSS,...</p>
        <div class="float-wrap">
            <h2>Will</h2>
            <p class="title-text">Today</p>
            <p class="title-date">2021-Now</p>
        </div>
            <p class="desc-text">End Web Developer</p>
            <p class="desc-subtext">Django,python</p>
        </section>
            <a href="http://www.naver.com">
                <img class="lena" src="./lena.png">
            </a>
    </div>

    <div>
        <footer>
          <p>copyright HYEON. All rights reserved</p>
        </footer>
    </div>

  </body>

</html>
'''
```

<h1>CSS</h1>
*  footer tag : p tag 를 사용한 것과 화면상 다를 게 없다. <br>
** text-align : center; - 중앙으로 정렬<br>
** float: left,right; - 붕붕떠다니면서 정렬<br>
** overflow: hidden; - float를 묶어준다<br>
** background-color: black; - 배경 색깔 <br>
** color: white; - 글자 색깔<br>
** font-size: 30px; - 글자 크기<br>
** font-weight: ligher - 글자 무게<br>
** font-style: italic - 글자 기울이기<br>
** line-height: 줄간간격 늘리기<br><br>

** class를 꾸밀 때 .클래스이름!<br>
** border: 두께/ 방식 /색깔 -상자그리기<br>
** border-bottom: border-top,right,left 상하좌우에 선이 생김<br>
** width : 상자 넓이 조절<br>
** height: 상자 높이 조절<br>
** margin-left, margin-right : 상자위치 조절<br>
** padding : 상자 크기와 글자 사이 거리 조절<br>
(꾸밀 태그 이름을 먼저 쓰고 중괄호를 열고 중괄호를 닫는다. 그리고 어떤 요소를 어떻게 꾸밀 건지 :과 ;으로 구분)<br>
(html color code에서 다양한 색깔을 고를 수 있다)<br>
<br>
** 상자의 외부넓이부터 Margin, Border,Padding,Content<br>
** box-shadow: 상자에 그림자 넣는 코드, 가로축/ 세로축/ 그림자의 흐림 정도/ 그림자의 퍼짐도/ 그림자의 색(rgba)(r,g,b,투명도)<br>
