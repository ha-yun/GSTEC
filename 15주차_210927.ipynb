```python

# gsweb\docker-compose.yml
version: "3.7"

services:

  nginx:
    image: nginx:1.19.5
    ports:
      - 80:80
    networks:
      - ha_network
    volumes:
      - /home/ubuntu/nginx.conf:/etc/nginx/nginx.conf
      - ha_static:/data/static/
      - ha_media:/data/media/


  django:
    image: ha_django:4
    networks:
      - ha_network
    volumes:
      - ha_static:/home/GSWEB/staticfiles/
      - ha_media:/home/GSWEB/media/
    secrets:
      - DJANGO_SECRET_KEY

  mariadb:
    image: mariadb:10.5
    networks:
      - ha_network
    volumes:
      - ha_mariadb:/var/lib/mysql
    environment:
      MARIADB_ROOT_PASSWORD: password1234
      MARIADB_DATABASE: django
      MARIADB_USER: django
      MARIADB_PASSWORD: password1234


networks:
  ha_network:


volumes:
  ha_static:
  ha_media:
  ha_mariadb:

secrets:
  DJANGO_SECRET_KEY:
    external: true
```
